```python
import pandas as pd
import numpy as np
import warnings
import matplotlib.pyplot as plt
import seaborn as sns 
df = pd.read_csv('downloads/archive/shopping_trends_updated.csv')
df.head()
```

```python
def get_age_group(Age):
    if Age >= 18 and Age <= 26:
        return 'Gen-Z'
    elif Age >=27 and Age <= 42:
        return 'Millenial'
    elif Age >=43 and Age <= 59:
        return 'Gen-X'
    elif Age >= 60:
        return 'Boomer'
df ['Age_group'] = df['Age'].apply(get_age_group)
```

```python
df['Age_group'].value_counts()
```

```python

df = pd.DataFrame ({'Age_group' : ['Gen-z', 'Millenial', 'Gen-x','Boomer'],'Age_count': [640, 1200, 640, 788]})
df
```

```python
df.groupby(['Age_group']).sum().plot(kind='pie', y='Age_count')
```
